```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## This plot verifies the inverse relationship between Temperature and humidity

```python
data= pd.read_csv("Urban Air Quality and Health Impact Dataset.csv")
data
```

```python
# selection of variables
x=data.datetime
y=data.temp
y1=data.humidity
```

```python
Temp=y.head(100)
```

```python
dates=np.arange(100)
```

```python
fig=plt.figure(figsize=(14,5))
plt.plot(dates,Temp, marker=".",label="temp")
plt.fill_between(dates, Temp - 0.5,Temp + 0.5,alpha=0.2)
plt.legend()
#plt.grid(True)
plt.xlabel("DateTime")
plt.ylabel("Temperature")
plt.title("Temperature trend for 30 Datetimes")
```

## HUMIDIDTY LEVELS

```python
hmdty=data.humidity.head(100)
```

```python
fig=plt.figure(figsize=(15,6))
plt.plot(hmdty, marker=".",label="humidity")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Humidity levels")
plt.title("Humidity Trend Analysis")
```

## correlation between temperature and humidity

```python
corr=np.corrcoef(Temp,hmdty)[0,1]
corr
```

## Correlation Plot Analysis

```python
fig, ax1=plt.subplots(figsize=(15,6))
ax1.plot(Temp.head(30), color="red", label="Temperature")
ax1.set_ylabel("Temperature")
ax2=ax1.twinx()
ax2.plot(hmdty.head(30), color="blue", label="Humidity")
ax2.set_ylabel("Humidity")
ax1.set_xlabel("Time")
ax2.legend(loc="center right")
ax1.legend(loc="center left")
ax1.text(0.05,0.95,f"correlation:{corr:.2f}", transform=ax1.transAxes,bbox=dict(facecolor='white', alpha=0.5),
        verticalalignment="center")
plt.title("Verification analysis between Temperature and Humidity")
```
